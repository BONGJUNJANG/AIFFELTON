```python
# 라이브러리 임포트 

import numpy as np
import pandas as pd


```

```python
bilstm_data = pd.read_csv('/content/drive/MyDrive/BILSTM_label.csv')
```

```python
dcnn_data = pd.read_csv('/content/drive/MyDrive/1DCNN_label.csv')
```

```python
voting = pd.DataFrame({'sum_0':bilstm_data['0'] + dcnn_data['0'],
                             'sum_1':bilstm_data['1'] + dcnn_data['1'],
                             'sum_2':bilstm_data['2'] + dcnn_data['2']})
```

```python
voting['pred_label'] = 0
```

```python
voting.loc[voting.max(axis=1) == voting['sum_0'], "pred_label"] = 0
voting.loc[voting.max(axis=1) == voting['sum_1'], "pred_label"] = 1
voting.loc[voting.max(axis=1) == voting['sum_2'], "pred_label"] = 2
```

```python
voting
```

```python
bilstm_data = bilstm_data.drop(labels='pred_labels',axis=1)
```

```python
ensamble_df = bilstm_data.join(voting['pred_label'],how='right') 
```

```python
ensamble_df
```

```python
ensamble_df.to_csv('bilstm_1dcnn_ensamble.csv')
```
