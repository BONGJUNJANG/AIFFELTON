```python
# 라이브러리 임포트 

import numpy as np
import pandas as pd


```

```python
bilstm_data = pd.read_csv('/content/drive/MyDrive/BILSTM_label.csv')
```

```python
RF_data = pd.read_csv('/content/drive/MyDrive/RF_Feature selection.csv')
```

```python
RF_data = RF_data.iloc[-len(bilstm_data):]
```

```python
RF_data = RF_data.drop(['Unnamed: 0'],axis=1)
RF_data = RF_data.drop(['RF_Predicted_label'],axis=1)
```

```python
bilstm_data
```

```python
RF_data
```

```python
index = bilstm_data.index
RF_data .index = index
```

```python
voting = pd.DataFrame({'sum_0':bilstm_data['0'] + RF_data['RF_Up day'],
                             'sum_1':bilstm_data['1'] + RF_data['RF_Down Day'],
                             'sum_2':bilstm_data['2'] + RF_data['RF_Stay']})
```

```python
voting['pred_label'] = 0
```

```python
voting.loc[voting.max(axis=1) == voting['sum_0'], "pred_label"] = 0
voting.loc[voting.max(axis=1) == voting['sum_1'], "pred_label"] = 1
voting.loc[voting.max(axis=1) == voting['sum_2'], "pred_label"] = 2
```

```python
voting
```

```python
bilstm_data = bilstm_data.drop(labels='pred_labels',axis=1)
```

```python
ensamble_df = bilstm_data.join(voting['pred_label'],how='right') 
```

```python
ensamble_df = ensamble_df.join(voting['sum_0'],how='right') 
ensamble_df = ensamble_df.join(voting['sum_1'],how='right') 
ensamble_df = ensamble_df.join(voting['sum_2'],how='right') 
ensamble_df = ensamble_df.drop(labels='0',axis=1)
ensamble_df = ensamble_df.drop(labels='1',axis=1)
ensamble_df = ensamble_df.drop(labels='2',axis=1)
```

```python
ensamble_df
```

```python
ensamble_df.to_csv('RF_Bilstm_Ensamble.csv')
```
